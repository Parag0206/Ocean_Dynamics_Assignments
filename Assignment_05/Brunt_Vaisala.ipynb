# Brunt Vaisala frequency

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy
import cartopy.crs as ccrs
```

```python
density_summer = xr.open_dataset("density_summer.nc", decode_times=False)
density_winter = xr.open_dataset("density_winter.nc", decode_times=False)
```

```python
density_summer = density_summer["I_an"]
density_winter = density_winter["I_an"]
```

```python
g = 9.8
rho_0 = 35.
```

## Bay of Bengal

```python
density_summer_bob = density_summer.sel(lon=["88."], method="nearest")
density_winter_bob = density_winter.sel(lon=["88."], method="nearest")
```

```python
density_summer_bob = density_summer_bob.sel(lat=slice(10, 20))
density_winter_bob = density_winter_bob.sel(lat=slice(10, 20))
```

```python
density_summer_bob.isel(time=0,lon=0).plot()
plt.ylim(4000,0)
```

```python
density_summer_bob
```

```python
N_summer_bob = density_summer_bob*0.0
for lat in range(len(density_summer_bob.lat.values)): 
    for depth in range(len(density_summer_bob.depth.values)-1):
        dz = density_summer_bob.depth[depth+1] - density_summer_bob.depth[depth]
        n = g/rho_0 * ((density_summer_bob[0,depth+1,lat,0]-density_summer_bob[0,depth,lat,0])/(dz.values))
        N_summer_bob[0,depth,lat,0] = np.sqrt(n)
```

```python
N_summer_bob.isel(time=0,lon=0).plot()
plt.ylim(1000,0)
plt.title("Brunt-Vaisala in JJAS, Bay of Bengal")
```

```python
N_winter_bob = density_winter_bob*0.0
for lat in range(len(density_winter_bob.lat.values)): 
    for depth in range(len(density_winter_bob.depth.values)-1):
        dz = density_winter_bob.depth[depth+1] - density_winter_bob.depth[depth]
        n = g/rho_0 * ((density_winter_bob[0,depth+1,lat,0]-density_winter_bob[0,depth,lat,0])/(dz.values))
        N_winter_bob[0,depth,lat,0] = np.sqrt(n)
```

```python
N_winter_bob.isel(time=0,lon=0).plot()
plt.ylim(1000,0)
plt.title("Brunt-Vaisala in DJF, Bay of Bengal")
```
